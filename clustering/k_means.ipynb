```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pval       = pd.read_csv('../data/BONF_PVAL_GEM2NET_UNIQLINK', sep = '\t', index_col = 0)
log_ratio  = pd.read_csv('../data/LOG_RATIO_GEM2NET_UNIQLINK', sep = '\t', index_col = 0)
matricePLM = pd.read_csv('../data/matricePLM_genes.txt', sep = '\t')
```

```python
df_log = log_ratio.copy()
df_log = df_log.dropna()

from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier

kmeans = KMeans(n_clusters = 10, random_state = 2, n_init = 10).fit(df_log)
df_log['label'] = kmeans.labels_
```

```python
nb_PLM = 0
for i in range(41,60):
    plt.figure(figsize =(8,8))
    for label in df_log['label'].unique():
        plt.plot(df_log.iloc[:,nb_PLM][df_log['label'] == label].values, df_log.iloc[:,i][df_log['label'] == label].values,
                '.', alpha = 0.7, label = label)
        plt.xlabel(f'gene {df_log.index[0]}')
        plt.ylabel(f'gene {df_log.index[i]}')
        plt.legend()
    
    plt.title('Clusters in different stress')
    plt.show()
```

```python
df_log_T = log_ratio.copy()
df_log_T = df_log_T.dropna()
df_log_T = df_log_T.T

kmeans = KMeans(n_clusters = 15, random_state = 2, n_init = 10).fit(df_log_T)
df_log_T['label'] = kmeans.labels_

df_log_T
```

```python
nb_PLM = 0
for i in range(1,10):
    plt.figure(figsize =(8,8))
    for label in df_log_T['label'].unique():
        plt.plot(df_log_T.iloc[:,nb_PLM][df_log_T['label'] == label].values, df_log_T.iloc[:,i][df_log_T['label'] == label].values,
                '.', alpha = 0.7, label = label)
        plt.xlabel(f'stress {df_log_T.index[0]}')
        plt.ylabel(f'stress {df_log_T.index[i]}')
        plt.legend()
    
    plt.title('Clusters in different genes')
    plt.show()
```

```python
df_pval  = pval.copy()
df_pval  = df_pval.dropna()
df_log_T = df_log_T.loc[:,df_pval.index]

df_log_T = df_log_T.applymap(lambda x: 1 if x > 0 else -1)

df_pval_T = df_pval.T
df_pval_T = df_pval_T.applymap(lambda x: 1 if x <= 0.05 else 0)

df_final = df_log_T*df_pval_T

df_final['label'] = kmeans.labels_
```

```python
for cluster in df_final['label'].unique():
    sns.clustermap(
        df_final.sort_values('label')[df_final['label'] == cluster].drop(columns= 'label'), 
        row_cluster=False,  
        col_cluster=False,
        cmap = 'viridis'
    )

    plt.show()
```
